## Daten laden

```python
import json
import requests
import pandas as pd

# URL of the JSON file
url = "https://raw.githubusercontent.com/pauline187/archiv/main/questionnaire_after_contact_interaction.json"

# Fetch JSON data from the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Parse JSON content
    print("JSON file successfully loaded")
else:
    print(f"Error: Unable to fetch data, status code {response.status_code}")
```

```python
with open('questionnaire.json', 'w', encoding='utf-8') as file:
    json.dump(data, file, indent=4)
print("JSON file saved locally as 'questionnaire.json'")
```

```python
with open('/content/questionnaire.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
rows = []
for item in data:
    question_id = item['id']
    question_type = item['type']
    question_text = item['question']

    for option in item['options']:
        rows.append({
            'id': question_id,
            'type': question_type,
            'question': question_text,
            'option': option['option']
        })

# Создаем DataFrame
df_table= pd.DataFrame(rows)
# Печатаем DataFrame
df_table.head(10)
```

```python
# Using only the singel select firstly
df_single_select = df_table[df_table['type'] == 'SINGLE_SELECT']
df_single_select.describe()
```

```python
# Loop to get answers for the head of the data set
# Leere Liste zum Speichern der Ergebnisse
responses = []
df_single_select5 = df_single_select.head()
# Schleife über jede Frage und Optionen
for index, row in df_single_select5.iterrows():
    prompt = f"You are a sales representative at a trade show, trying to connect with people. You have a conversation with someone (pls imagen the whole conversation) and then you fill out a questionnaire in an app, trying to remember as much as possible about the interaction you just had. The question you answer is {row['question']} the options are {row['option']}. Give me only the context why you chose a option not the option you chose."


    # Antwort vom Modell abrufen
    response = model.generate_content(prompt)

    answer = response.text

    # Antwort speichern
    responses.append({
        'question': row['question'],
        'options': row['option'],
        'answer': answer
    })

# DataFrame aus den Ergebnissen erstellen
df_answers = pd.DataFrame(responses)
print(df_answers)


```

```python
# Loop to get more answers
import time
# Empty list to store the results
responses = []


# Loop through each row and generate a response
for index, row in df_single_select.iterrows():
    prompt = f"""
    You are a sales representative at a trade show, trying to connect with people.
    You have a conversation with someone (please imagine the whole conversation)
    and then you fill out a questionnaire in an app, trying to remember as much
    as possible about the interaction you just had.

    The question you answer is: "{row['question']}"
    The options are: "{row['option']}".

    Give me only the context of why you chose an option, not the option itself.
    """

    try:
        # Call the Gemini API to generate content
        response = model.generate_content(prompt)

        # Safely extract the answer text
        if response and response.text:
            answer = response.text.strip()
        else:
            answer = "Error: No response generated"

    except Exception as e:
        answer = f"Error generating response: {e}"

    # Save the question, options, and generated answer
    responses.append({
        'question': row['question'],
        'options': row['option'],
        'answer': answer
    })
    time.sleep(2)

# Convert the results to a DataFrame
df_answers = pd.DataFrame(responses)

# Display the generated answers
print(df_answers)

# Save the results to a CSV file
#df_answers.to_csv('generated_answers.csv', sep='>', index=False, encoding='utf-8')

#print("Generated answers saved to 'generated_answers.csv'")
```
